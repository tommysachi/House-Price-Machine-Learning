# Machine Learning
- Linear Regression, Ridge, Lasso, Elastic Net

```python
from sklearn.model_selection import train_test_split
```

```python
dfML = df1[['price','bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront','view','condition','grade','yr_built','year_renovated_recat']]
```

```python
dfML
```

```python
X = dfML.drop(columns='price')
y = dfML['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = .80,random_state=42)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

# Linear Regression

```python
Model_1 = LinearRegression(fit_intercept=False)
```

```python
Model_1.fit(X_train,y_train)
```

```python
def Eva_Matrix(Model,X,y_true):
    y_pred = Model.predict(X)
    R2 = r2_score(y_true,y_pred)
    MAE = mean_absolute_error(y_true,y_pred)
    MSE = mean_squared_error(y_true,y_pred)
    RMSE = np.sqrt(MSE)
    return R2, MAE, MSE, RMSE
```

```python
r2_train_1, MAE_train_1, MSE_train_1, RMSE_train_1 = Eva_Matrix(Model_1,X_train,y_train)
```

```python
r2_test_1, MAE_test_1, MSE_test_1, RMSE_test_1 = Eva_Matrix(Model_1,X_test,y_test)
```

```python
data1 = {
    "Training" : [r2_train_1,MAE_train_1,MSE_train_1,RMSE_train_1],
    "Testing" : [r2_test_1,MAE_test_1,MSE_test_1,RMSE_test_1]
}
pd.DataFrame(data=data1,index=['R2','MAE','MSE','RMSE'])
```

# Ridge

```python
Model_R = Ridge(fit_intercept=False)
```

```python
Model_R.fit(X_train,y_train)
```

```python
r2_train_R, MAE_train_R, MSE_train_R, RMSE_train_R = Eva_Matrix(Model_R,X_train,y_train)
r2_test_R, MAE_test_R, MSE_test_R, RMSE_test_R = Eva_Matrix(Model_R,X_test,y_test)
```

```python
dataR = {
    "Training" : [r2_train_R, MAE_train_R, MSE_train_R, RMSE_train_R],
    "Testing" : [r2_test_R, MAE_test_R, MSE_test_R, RMSE_test_R]
}
pd.DataFrame(data=dataR,index=['R2','MAE','MSE','RMSE'])
```
